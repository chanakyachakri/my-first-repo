What is the difference between hardware and software? Give one example of each.

Software is a set of Computer programs like code, algorithms, data structures, software applications that instructs computers what to do
Example: OS, Choorme, Tools like MS-Office, apps like whatsapp, twitter, and etc.

Hardware is a physical components of a computer system like CPU, Monitor, Mouse, Keyboard, RAM and etc.
